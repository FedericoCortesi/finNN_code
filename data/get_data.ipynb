```python
import wrds
import pandas as pd
```

```python
db = wrds.Connection()
```

```python
start_date = '2000-01-01'
end_date = '2024-12-31'

sp500_daily_query = f"""
SELECT
    s.permno, 
    s.date, 
    s.ret, 
    s.prc AS close, 
    s.vol, 
    s.openprc AS open, 
    s.askhi AS high, 
    s.bidlo AS low, 
    s.cfacpr 
FROM
    crsp.dsf AS s
INNER JOIN
    crsp.msp500list AS m ON s.permno = m.permno
WHERE
    s.date BETWEEN m.start AND m.ending
    AND s.date BETWEEN '{start_date}' AND '{end_date}'
ORDER BY
    s.permno, 
    s.date;
"""

sp500_daily_data = db.raw_sql(sp500_daily_query)
sp500_daily_data
```

```python
sp500_daily_data.to_parquet('sp500_daily_data.parquet')
```

```python
db.close()
print("\nWRDS connection closed")
```
